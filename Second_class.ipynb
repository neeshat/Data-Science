# 3. Calculate the area and perimeter of a rectangle with length 7 and width 5.

```python
length = 7
width = 5
area = length * width
perimeter = 2 * (length + width)

print("The area is: ",area)
print("The perimeter is: ",perimeter)
```

# 8. Create two sets of colors and find their union and intersection

```python
color_set1 = {"red", "green", "blue"}
color_set2 = {"yellow", "green", "purple"}

union_set = color_set1 | color_set2
intersection_set = color_set1 & color_set2

print("Union set: ",union_set)
print("Intersection set: ",intersection_set)
```
